```python
#Jupyter notebook for analyzing the processed data, understanding patterns, and exploring trends before modeling.
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Set display and plot style
pd.set_option("display.max_columns", None)
sns.set(style="whitegrid")
```

```python
# Load processed data
df = pd.read_csv("../data/processed/historical_data_2025-07-17_01-05-15.csv")
df["date"] = pd.to_datetime(df["date"])


# view basic info
print(df.info())
print(df.head())
```

```python
# Check correlation between temperature and energy consumption
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x="TMAX", y="energy_consumption", hue="city")
plt.title("TMAX vs Energy Consumption")
plt.xlabel("Max Temperature ()")
plt.ylabel("Energy Consumption (MWh)")
plt.show()
```

```python
# Time-series trend for a city
city = "New York"
df_city = df[df["city"] == city]

plt.figure(figsize=(12, 6))
plt.plot(df_city["date"],  df_city["energy_consumption"], label="Energy")
plt.plot(df_city["date"], df_city["TMAX"], label="TMAX", alpha=0.6)
plt.title(f"Energy and Temperature Over Time  {city}")
plt.xlabel("Date")
plt.legend()
plt.tight_layout()
plt.show()

#If energy usage follows temperature trends

#If data looks noisy, seasonal, or lagged

#If there's enough variance for ML training
```
